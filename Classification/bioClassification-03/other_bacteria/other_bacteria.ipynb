```python
from Bio import SeqIO

# FASTA dosyalarının yolları
fasta_files = ["uniprotkb_Bacillus_subtilis.fasta", "uniprotkb_Clostridium_difficile.fasta", "uniprotkb_Escherichia_coli.fasta",
               "uniprotkb_Klebsiella_pneumoniae.fasta", "uniprotkb_Lactobacillus_acidophilus.fasta", "uniprotkb_Mycobacterium_tuberculosis.fasta", "uniprotkb_Staphylococcus_aureus.fasta"]

# Çıktı dosyasının adı
output_file = "combined_bacteria.fasta"

with open(output_file, "w") as outfile:
    # Her bir FASTA dosyası okunuyor ve yazılıyor
    for fasta_file in fasta_files:
        with open(fasta_file, "r") as infile:
            outfile.write(infile.read())

print(f"FASTA dosyaları başarıyla birleştirildi ve {output_file} dosyasına kaydedildi.")
```

```python
from Bio import SeqIO
import pandas as pd

file_path = "combined_bacteria.fasta"  # Birleştirilmiş FASTA dosyasının yolu
def read_fasta(file_path):
    sequences = []
    for record in SeqIO.parse(file_path, "fasta"):
        sequences.append({"ID": record.id, "Sequence": str(record.seq)})
    return pd.DataFrame(sequences)

# FASTA dosyasının okunması
fasta_file = "combined_bacteria.fasta"
data = read_fasta(fasta_file)

# Veriyi CSV formatında kaydediyoruz
csv_file = "other_bacteria_sequences.csv"
data.to_csv(csv_file, index=False)

print(f"FASTA dosyası başarıyla CSV'ye dönüştürüldü: {csv_file}")
```
