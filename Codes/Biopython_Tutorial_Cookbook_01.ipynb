**Transcription**

\begin{split}\begin{gathered}
    \text{DNA coding strand (aka Crick strand, strand } +1 \text{)} \\
    \text{5'} \qquad \texttt{ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG} \qquad \text{3'} \\
    \texttt{|||||||||||||||||||||||||||||||||||||||} \\
    \text{3'} \qquad \texttt{TACCGGTAACATTACCCGGCGACTTTCCCACGGGCTATC} \qquad \text{5'} \\
    \text{DNA template strand (aka Watson strand, strand } -1 \text{)}
\end{gathered}\end{split}

\begin{split}\begin{gathered}
    \text{5'} \qquad \texttt{AUGGCCAUUGUAAUGGGCCGCUGAAAGGGUGCCCGAUAG} \qquad \text{3'} \\
    \text{Single-stranded messenger RNA}
\end{gathered}\end{split}

Gerçek biyolojik transkripsiyon işlemi, mRNA'yı vermek için ters bir tamamlama (TCAG → CUGA) yaparak şablon şeritten çalışır. Bununla birlikte, Biopython'da ve genel olarak biyoinformatikte, genellikle doğrudan kodlama ipliği ile çalışırız çünkü bu, mRNA dizisini sadece T → U değiştirerek alabileceğimiz anlamına gelir.

```python
from Bio.Seq import Seq
coding_dna = Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG")
coding_dna
```

```python
template_dna = coding_dna.reverse_complement()
template_dna
```

```python
coding_dna
```

```python
messenger_rna = coding_dna.transcribe()
messenger_rna
```

```python
template_dna.reverse_complement().transcribe()
```

```python
# Seq nesnesi ayrıca mRNA'dan DNA'nın kodlama ipliğine gitmek için bir geri transkripsiyon yöntemi içerir. Yine, bu basit bir U → T ikamesidir:
from Bio.Seq import Seq
messenger_rna = Seq("AUGGCCAUUGUAAUGGGCCGCUGAAAGGGUGCCCGAUAG")
messenger_rna
```

```python
messenger_rna.back_transcribe()
```

**MutableSeq objects**

Normal Python dizesi gibi, Seq nesnesi de "salt okunur" veya Python terminolojisinde değişmezdir. Seq nesnesinin bir dize gibi davranmasını istemenin yanı sıra, bu aynı zamanda yararlı bir varsayılandır çünkü birçok biyolojik uygulamada dizi verilerinizi değiştirmediğinizden emin olmak istersiniz

Ancak, bunu değiştirilebilir bir diziye (MutableSeq nesnesi) dönüştürebilir ve onunla istediğiniz hemen hemen her şeyi yapabilirsiniz

Alternatif olarak, bir dizeden doğrudan bir MutableSeq nesnesi oluşturabilirsiniz

```python
from Bio.Seq import Seq
my_seq = Seq("GCCATTGTAATGGGCCGCTGAAAGGGTGCCCGA")
```

```python
my_seq[5] = "G"
```

```python
from Bio.Seq import MutableSeq
mutable_seq = MutableSeq(my_seq)
mutable_seq
```

```python
from Bio.Seq import MutableSeq
mutable_seq = MutableSeq("GCCATTGTAATGGGCCGCTGAAAGGGTGCCCGA")
```

```python
mutable_seq
```

```python
mutable_seq[5] = "C"
mutable_seq
```

```python
mutable_seq.remove("T")
mutable_seq
```

```python
mutable_seq.reverse()
mutable_seq
```

```python
from Bio.Seq import Seq
new_seq = Seq(mutable_seq)
new_seq
```

Finding subsequences

Dizi nesneleri, düz dize nesnelerinde karşılık gelen yöntemlerle aynı işlevi yerine getiren "find", "rfind", "index" ve "rindex" yöntemlerine sahiptir. Tek fark, alt dizinin bir dize, "bytes", "bytearray", "Seq" veya "MutableSeq" nesnesi olabilmesidir

```python
from Bio.Seq import Seq, MutableSeq
seq = Seq("GCCATTGTAATGGGCCGCTGAAAGGGTGCCCGA")
seq.index("ATGGGCCGC")
```

```python
seq.index(b"ATGGGCCGC")
```

```python
seq.index(bytearray(b"ATGGGCCGC"))
```

```python
seq.index(Seq("ATGGGCCGC"))
```

```python
seq.index(MutableSeq("ATGGGCCGC"))
```

```python
seq.index("ACTG")
```

```python
seq.find("ACTG")
```

```python
seq.find("CC")
```

```python
seq.rfind("CC")
```

```python
for index, sub in seq.search(["CC", "GGG", "CC"]):
    print(index, sub)
```
